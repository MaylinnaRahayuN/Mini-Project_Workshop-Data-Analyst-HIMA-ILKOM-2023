## **Import Library**

````python
import pandas as pd
import matplotlib.pyplot as plt
````

## **Load Data**

````python
# Load the data from the CSV file
df = pd.read_csv('/content/Telco Customer Churn.csv')
````

````python
print(df) #Display the loaded data
````

## **1. Generate descriptive statistic for numerical (continues) columns**


````python
# Generate descriptive statistics for numerical columns
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']  #membuat sebuah list
numerical_stats = df[numerical_cols].describe() #menghitung statistik
print("Descriptive statistics for numerical columns:")
print(numerical_stats)
````

````python
import matplotlib.pyplot as plt

# Generate descriptive statistics for numerical columns
numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
numerical_stats = df[numerical_cols].describe()

# Create bar plot
fig, ax = plt.subplots()
numerical_stats.plot(kind='bar', ax=ax)
ax.set_title('Descriptive Statistics for Numerical Columns')
ax.set_ylabel('Values')

# Show the plot
plt.show()

````

## **2. Generate descriptive statistic for categorical columns**


````python
# Generate descriptive statistics for categorical columns
categorical_cols = ['gender', 'Dependents', 'Contract', 'Churn']
categorical_stats = df[categorical_cols].describe() #hitung statistik
print("\nDescriptive statistics for categorical columns:")
print(categorical_stats)
````

````python
import matplotlib.pyplot as plt

# Generate descriptive statistics for categorical columns
categorical_cols = ['gender', 'Dependents', 'Contract', 'Churn']
categorical_stats = df[categorical_cols].describe()

# Create bar plot
fig, ax = plt.subplots()
categorical_stats.transpose().plot(kind='bar', ax=ax)
ax.set_title('Descriptive Statistics for Categorical Columns')
ax.set_ylabel('Count')

# Show the plot
plt.show()
````

## **5. Create a data visualization of comparison of churn and non-churn customers based on gender, dependents, and contracts**

````python
# Create data visualizations
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Comparison of churn and non-churn customers based on gender
gender_churn_counts = df.groupby(['gender', 'Churn']).size().unstack()
gender_churn_counts.plot(kind='bar', stacked=True, ax=axes[0])
axes[0].set_title('Churn vs Non-churn Customers by Gender')
axes[0].set_ylabel('Count')

# Comparison of churn and non-churn customers based on dependents
dependents_churn_counts = df.groupby(['Dependents', 'Churn']).size().unstack()
dependents_churn_counts.plot(kind='bar', stacked=True, ax=axes[1])
axes[1].set_title('Churn vs Non-churn Customers by Dependents')
axes[1].set_ylabel('Count')

# Comparison of churn and non-churn customers based on contracts
contracts_churn_counts = df.groupby(['Contract', 'Churn']).size().unstack()
contracts_churn_counts.plot(kind='bar', stacked=True, ax=axes[2])
axes[2].set_title('Churn vs Non-churn Customers by Contract')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
````

## **6. Check if there are any outliers in tenure data for churn and non-churn customers**


````python
# Check for outliers in tenure data for churn and non-churn customers
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

df.boxplot(column='tenure', by='Churn', ax=axes[0])
axes[0].set_title('Churn vs Non-churn Customers - Tenure')
axes[0].set_ylabel('Tenure')

df.boxplot(column='tenure', by='Churn', showfliers=False, ax=axes[1])
axes[1].set_title('Churn vs Non-churn Customers - Tenure (without outliers)')
axes[1].set_ylabel('Tenure')

plt.tight_layout()
plt.show()
````
